## **GERAÇÃO DE DADOS:**

```python
import random                                                     
import csv                                                        
from num2words import num2words                                   
from faker import Faker                                           
import matplotlib.pyplot as plt                                   
from wordcloud import WordCloud                                   
from collections import Counter                                   
                                                                  
                                                                  
fake = Faker("pt-BR")                                             
name_list = [fake.name() for num in range(1, 11)]                 
final_list = {name: random.randint(1, 100) for name in name_list}
```

## **GRAVAÇÃO DE DADOS:**

```python
                                                           
data_file = open('final_list.csv', 'w', encoding='utf-8')  
                                                           
write = csv.writer(data_file)                              
                                                           
for key in final_list:                                     
    write.writerow([key, final_list[key]]) 

data_file.close()                 
                                                           
```

## **LEITURA DE DADOS:**

```python
extract_data1 = []
process_data = []
with open('final_list.csv', 'r') as final_data:
    final_data_csv = csv.reader(final_data, delimiter=',')
    extract_data1.extend(final_data_csv)

    for data in extract_data1:
        if data:
            process_data.append(data)

    grade_list = [data[1] for data in process_data]
    grade_list.sort(key = lambda x: int(x.split('-')[0]))


num_list = [num2words(num, lang="pt-br") for num in grade_list]
```

## **HISTOGRAMA:**

```python
plt.title("Histograma das Pontuações")                              
plt.xlabel("Pontuação")                                             
plt.ylabel("Probabilidade")                                         
plt.hist(grade_list, 10, density=True, rwidth=0.9, color="Blue")    
plt.show()                                                          
```

## **NUVEM DE PALAVRAS:**

```python
num_dict = Counter(num_list)                                                        
word_cloud = WordCloud(width=900, height=550).generate_from_frequencies(num_dict)   
                                                                                    
plt.figure()                                                                        
plt.imshow(word_cloud)                                                              
plt.axis('off')                                                                     
plt.savefig('word_cloud.png', bbox_inches='tight')                                  
plt.show()                                                                          
plt.close()                                                                         
```

###**a) Qual a importância de manipular arquivos de texto?**
R- A manipulação de arquivos de texto permite tanto ao programador quanto ao usuário final do programa a extração, análise e interpretação de dados de diversas fontes, abrindo assim um leque enorme de possibilidades sobre a fonte dos dados e o formato bruto desses dados, agilizando também o trabalho do usuário final que receberá os dados já tratados para serem analisados.  

###*b) Qual a importância de visualizar os dados em um histograma?*
R- A visualização de dados num histograma é importante pois nos mostra de forma mais clara, prática e rápida como determinado dado está distribuído, sua amplitude, simetria e sua variação em torno de um ponto central.

###*c) O que significa a visualização por nuvem de palavras?*
R- A visualização por nuvens de palavras significa ver a frequência com que certas palavras estão sendo utilizadas em determinada fonte de dados, essa frequência vai definir o tamanho da fonte com que a palavra será escrita, sendo assim, quanto maior for a frequência de utilização da palavra, maior será o seu tamanho na nuvem
